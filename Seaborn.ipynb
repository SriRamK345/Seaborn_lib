```python
import seaborn as sns
```

```python
x = ["jan", "feb", "march", "april", "may"]
sales = [1000, 1200, 900, 400, 500]
```

## Bar chart

```python
sns.barplot(x)
```

```python
import matplotlib.pyplot as plt
```

```python
colr = sns.color_palette("husl", len(x))

plt.bar(x, sales, color=colr)
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=df)
```

```python
sns.scatterplot(x="total_bill", y="tip", data=df, hue="total_bill", palette="husl")
```

```python
sns.lineplot(x="total_bill", y="tip", data=df, hue="day", palette="husl")
```

```python
sns.lineplot(x="total_bill", y="tip", data=df, palette="husl")
```

## Sub_plot

```python
sns.set_theme(style="darkgrid")
fig, ax = plt.subplots(1,2)
sns.lineplot(x="total_bill", y="tip", data=df, palette="husl")
```

```python
sns.boxplot(y="total_bill", x="day", data=df, palette="husl")
plt.show()
```

```python
sns.histplot(x="total_bill", data=df, palette="husl")
plt.show()
```

```python
sns.displot(x="total_bill", data=df, palette="husl")
plt.show()
```

```python
sns.pairplot(df, hue = "day")
plt.show()
```

## kde plot

-  Kernel Density Estimate (KDE) plot visualizes the distribution of a dataset's observations.

```python
sns.kdeplot(df["total_bill"])
plt.show()
```

```python
sns.kdeplot(x = "total_bill", y = "tip", data = df)
plt.show()
```

## Pie chart

```python
plt.pie(sales, labels=x, autopct="%1.1f%%")
plt.show()
```
